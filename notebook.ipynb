Una cooperativa agrícola en Guatemala los ha contratado para automatizar la detección de enfermedades

combinados con un SVM) porque "son más ligeros y no requieren GPUs". Ustedes, como directores del

1. Redacte una justificación técnica (máximo 2 párrafos) dirigida al gerente de TI explicando por qué
SIFT/HOG fallarían en un entorno agrícola real frente a variaciones de luz, ángulos y oclusiones, y
cómo los "Mapas de Características" de una red profunda solucionan la variabilidad semántica

Los algoritmos tradicionales como SIFT y HOG dependen de la extracción manual de características geometricas y gradientes de intensidad que resultan extremadamente fragiles ante las condiciones impredecibles de un entorno agricola real ya que en el campo las variaciones drasticas de iluminación natural, los diferentes angulos de captura de las hojas y las oclusiones parciales por ramas o sombras rompen los patrones que estos descriptores intentan detectar lo que causa una alta tasa de falsos negativos en la deteccion de enfermedades, al ser metodos que no aprenden de forma jerarquica y carecen de la robustez necesaria para generalizar mas alla de condiciones controladas de laboratorio. 

Por otro lado el uso de Deep Learning mediante redes neuronales convolucionales permite generar mapas de caracteristicas que aprenden de forma automatica y jerarquica a ignorar el ruido visual y centrarse en la variabilidad semantica de la enfermedad, estas redes extraen desde texturas basicas hasta patrones complejos de necrosis o manchas que son invariantes a la rotación y a los cambios de luz, permitiendo que el modelo comprenda que es una enfermedad de mango sin importar como este iluminada o posicionada la hoja. Esto garantiza una precisión y confiabilidad muy superiores en el diagnóstico automatizado justificando la inversión en hardware.

2. Si tuviera que elegir estrictamente entre usar la arquitectura AlexNet o VGG-16 para extraer
características de las hojas, ¿cuál elegiría y por qué? Justifique su respuesta basándote en la
modularidad, la no-linealidad (ReLU) y la capacidad de abstracción de capas profundas.

Para este proyecto yo elegiría la arquitectura VGG-16 ya que AlexNet la VGG-16 destaca por su alta modularidad ya que utiliza bloques repetitivos de filtros pequeños de 3x3 con stride de 1 y padding de 1 lo que permite apilar muchas mas capas de forma estandarizada y eficiente, esta mayor profundidad implica una mayor cantidad de funciones de activación Relu intercaladas, lo que otorga a la red una capacidad superior para modelar las complejas variaciones de textura y color presentes en las enfermedades de las hojas de mango, por otro lado mientras que AlexNet tiene una estructura mas rígida y superficial la profundidad de VGG-16 permite una abstracción semantica mucho mejor descomponiendo la imagen en niveles jerarquicos que van desde bordes simples hasta patrones patologicos detallados para asi garantizar una extracción de características mucho mas robusta para el entorno agricola.
